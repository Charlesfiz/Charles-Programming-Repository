```python
# Stopwatch

# define global variables
import simplegui

t = 0
num_success = 0
num_total_stop = 0

# define helper function format that converts time in tenths of seconds into formatted string A:BC.D
def format(t):
    A = t // 600 # 0 - 9
    global sec
    sec = t // 10
    B = (sec % 60) // 10 # 0 - 5
    C = (sec % 60) % 10 # 0 - 9
    D = t % 10 # 0 - 9
    return str(A) + ':' + str(B) + str(C) + '.' + str(D)
        
# define event handlers for buttons; "Start", "Stop", "Reset"
def Start():
    timer.start()

def Stop():
    timer.stop()
    global num_total_stop
    global num_success
    num_total_stop += 1
    if t % 10 == 0:
        num_success += 1
    
def Reset():
    global t
    global num_total_stop
    global num_success
    timer.stop()
    t = 0
    num_total_stop = 0
    num_success = 0
        
# define event handler for timer with 0.1 sec interval
def Tick():
    global t
    t += 1
    
# define draw handler
def Draw(canvas):
    canvas.draw_text(format(t), [65, 100], 30, 'White')
    canvas.draw_text(str(num_success) + '/' + str(num_total_stop), [150, 40], 20, 'Yellow')
    
# create frame
frame = simplegui.create_frame('StopWatch', 200, 200)

# register event handlers
timer = simplegui.create_timer(100, Tick)
frame.set_draw_handler(Draw)
frame.add_button('Start', Start, 100)
frame.add_button('Stop', Stop, 100)
frame.add_button('Reset', Reset, 100)

# start frame
frame.start()

# http://www.codeskulptor.org/#user45_tKkLLKVa5l_0.py
```
