```python
# Guess the Number
import simplegui
import random

remaining_guess_number = 7
num_range = 100

# helper function to start and restart the game
def new_game():
    # initialize global variables used in your code here
    global secret_number
    global num_range
    global remaining_guess_number
    secret_number = random.randrange(num_range)
    print "\nNew game. Range is 0 to " + str(num_range)
    print "Number of remaining guesses is " + str(remaining_guess_number)
    
# define event handlers for control panel
def range100():
    # button that changes the range to [0,100) and starts a new game 
    global secret_number
    global num_range
    global remaining_guess_number
    num_range = 100
    secret_number = random.randrange(num_range)
    remaining_guess_number = 7
    new_game()

def range1000():
    # button that changes the range to [0,1000) and starts a new game     
    global secret_number
    global num_range
    global remaining_guess_number
    num_range = 1000
    secret_number = random.randrange(num_range)
    remaining_guess_number = 10
    new_game()
    
def input_guess(guess):
    # main game logic
    print '\nGuess was ' + guess
    global remaining_guess_number
    global num_range
    remaining_guess_number -= 1
    if remaining_guess_number > -1:    
        if int(guess) > (num_range - 1):
            print "Number of remaining guesses is " + str(remaining_guess_number)
            print "Out of guesses"
        elif int(guess) < 0:
            print "Number of remaining guesses is " + str(remaining_guess_number)
            print "Out of guesses"
        elif int(guess) > secret_number:
            print "Number of remaining guesses is " + str(remaining_guess_number)
            print "Lower"
        elif int(guess) < secret_number:
            print "Number of remaining guesses is " + str(remaining_guess_number)
            print "Higher"
        else:
            print "Correct"
            if num_range == 100:
                remaining_guess_number = 7
                new_game()
            else:
                remaining_guess_number = 10
                new_game()
    else:
        if num_range == 100:
            remaining_guess_number = 7
            new_game()
        else:
            remaining_guess_number = 10
            new_game()

# create frame
frame = simplegui.create_frame("GTN", 200, 200)

# register event handlers for control elements and start frame
inp = frame.add_input("input_guess", input_guess, 100)
button1 = frame.add_button("Range is [0,100)", range100, 150)
button2 = frame.add_button("Range is [0,1000)", range1000, 150)

# call new_game
frame.start()
new_game()

# http://www.codeskulptor.org/#user45_OpWT6QG0JZ_4.py
```
