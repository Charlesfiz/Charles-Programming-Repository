```python
# implementation of card game - Memory

import simplegui
import random

WIDTH = 800
HEIGHT = 100
count = 8
lst_1 = []
lst_2 = []
rlst = []
exposed = [False, False, False, False, False, False, False, False, False, False, False, False, False, False, False, False]
state = 0
turn = 0
c = 0
d = 0

while count > 0:
    r = random.randrange(0,8)
    if r not in lst_1:
        lst_1.append(r)
        count -= 1
    else:
        continue
        
count = 8
        
while count > 0:
    r = random.randrange(0,8)
    if r not in lst_2:
        lst_2.append(r)
        count -= 1
    else:
        continue
        
rlst = lst_1 + lst_2

# helper function to initialize globals
def new_game():
    global rlst, state, turn, exposed
    exposed = [False, False, False, False, False, False, False, False, False, False, False, False, False, False, False, False]
    random.shuffle(rlst)
    state = 0
    turn = 0
    label.set_text("Turns = " + str(turn))
     
# define event handlers
def mouseclick(pos):
    # add game state logic here
    global state, a, b, turn, c, d
    position = list(pos)
    if state == 0:
        a = position[0] // 50
        exposed[a] = True
        state = 1
    elif state == 1:
        b = position[0] // 50
        exposed[b] = True
        state = 2
    elif state == 2:
        turn += 1
        label.set_text("Turns = " + str(turn))
        if rlst[a] != rlst[b]:
            exposed[a] = False
            exposed[b] = False
        elif rlst[a] == rlst[b]:
            c = a
            d = b
        a = position[0] // 50
        exposed[a] = True
        state = 1
         
# cards are logically 50x100 pixels in size    
def draw(canvas):
    for r in range(len(rlst)):
        card_pos = [20 + r*50, HEIGHT/1.6] 
        rect_pos = [[0 + r*50, 0], [50 + r*50, 0], [50 + r*50, 100], [0 + r*50, 100]]
        
        # if exposed, make the number visible; otherwise, hide the card
        if exposed[r] == True:
            canvas.draw_text(str(rlst[r]), [card_pos[0], card_pos[1]], 35, 'White')
        else:
            canvas.draw_polygon([rect_pos[0], rect_pos[1], rect_pos[2], rect_pos[3]], 1, 'Red', 'Green')
                
# create frame and add a button and labels
frame = simplegui.create_frame("Memory", WIDTH, HEIGHT)
frame.add_button("Reset", new_game)
label = frame.add_label("Turns = 0")

# register event handlers
frame.set_mouseclick_handler(mouseclick)
frame.set_draw_handler(draw)

# get things rolling
new_game()
frame.start()

# http://www.codeskulptor.org/#user45_duQD9x9Cwe_0.py
```
